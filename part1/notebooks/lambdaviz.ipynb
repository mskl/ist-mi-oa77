```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from io import StringIO 
```

```python
data1 = """
lambda&changes&mean deviation
-3&79&0.1257
-2&79&0.8242
-1&79&2.1958
0&79&3.6826
1&79&5.6317
2&79&10.9041
3&79&15.3304
"""
df1 = pd.read_csv(StringIO(data1), sep="&", index_col=0)

data2 = """
lambda&changes&mean deviation
-3&10&0.0075
-2&8&0.0747
-1&11&0.7021
0&5&2.8876
1&4&5.3689
2&3&12.5914
3&1&16.2266
"""
df2 = pd.read_csv(StringIO(data2), sep="&", index_col=0)


data3 = """
lambda&changes&mean deviation
-3&13&0.0107
-2&12&0.1055
-1&14&0.8863
0&11&2.8732
1&6&5.4361
2&2&13.0273
3&2&16.0463
"""
df3 = pd.read_csv(StringIO(data3), sep="&", index_col=0)
```

```python
df1.columns
```

```python
mdf = df1[["mean deviation"]].rename(columns={"mean deviation": "task 1"})
mdf["task 2"] = df2["mean deviation"]
mdf["task 3"] = df3["mean deviation"]
mdf.index = pow(10.0, mdf.index)

cdf = df1[["changes"]].rename(columns={"changes": "task 1"})
cdf["task 2"] = df2["changes"]
cdf["task 3"] = df3["changes"]
cdf.index = pow(10.0, cdf.index)
```

```python
fig, ax = plt.subplots(figsize=(14, 7))
mdf.plot(ax=ax, color=["tab:blue", "tab:orange", "tab:green"])
#ax.set_title("Waypoint mean deviation for Tasks 1 to 3.")
ax.set_ylabel("mean deviation")
ax.set_xscale("log")
plt.tight_layout()
plt.savefig('task_1_2_3_waypoint_mean_dev.pdf')  
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(14, 7))
cdf[["task 2", "task 3"]].plot(ax=ax, kind="bar", color=["tab:orange", "tab:green"])
#ax.set_title("Control signal changes for Tasks 2 and 3.")
ax.set_ylabel("number of changes")
plt.tight_layout()
plt.savefig('task_2_3_signal_changes.pdf')  
plt.show()
```
